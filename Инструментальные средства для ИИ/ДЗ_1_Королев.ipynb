## Многомерная линейная регрессия из sklearn

Применим многомерную регрессию из sklearn для стандартного датасета

```python
from sklearn.datasets import make_regression  # Линейная регрессия — это метод, который используется для предсказания значений по значениям
                                              # Например, если мы хотим предсказать цену дома на основе его площади, количества комнат и тд, линейная регрессия помогает установить связь между этими переменными
X, y = make_regression(n_samples = 10000)  # Искусственно создаем набор данных с х = 10000 объектов и у = 100 (по умолчанию) признаков
                                           # х - матрица признаков (например, площадь, кол-во комнат), у - целевая переменная (например, цена)
print(X.shape, y.shape)
```

У нас 10000 объектов и 100 признаков. Для начала решим задачу аналитически "из коробки" (то есть сразу после запуска)

```python
from sklearn.linear_model import LinearRegression # импортируем класс, чтобы позже обучить модель на наших данных
from sklearn.metrics import mean_squared_error    # для оценки качества модели

reg = LinearRegression().fit(X, y)                # обучаем модель с помощью класса "LinearRegression" методом "fit" на данных х и у
print("Среднеквадратическая ошибка: ", mean_squared_error(y, reg.predict(X)))       # оцениваем качество модели (чем меньше ошибка, тем лучше)
```

***Задание 2 (0.5 балла).*** Подберите гиперпараметры в методе градиентного спуска так, чтобы значение MSE было близко к значению MSE, полученному при обучении LinearRegression.

Для приближения значения MSE при градиентном спуске с MSE аналитического метода, необходимо поэксперементировать с несколькими параметрами:

- Скорость обучения

По началу использовался массив из 3-5 значений alpha, но все они были недостаточно хороши, поэтому была написана логарифмическая последовательность значений alpha с большим диапозоном.

- Количество итераций

Для увеличения времени на обучение, параметр mmax_iter был увеличен до 1000
- Толерантность

Для улучшения оптимизации был установлен параметр tol на 1e-10, чтобы привести к лучшему совпадению ошибки

```python
# Целевое значение MSE
target_mse = 1.2829895487030765e-25

# Генерация значений alpha от 1e-6 до 1e-35 с логарифмическим шагом
alpha_values = np.logspace(-6, -35, num=30)  # 30 значений между 1e-6 и 1e-35

best_mse_diff = float('inf')  # Для отслеживания ближайшего значения к целевому MSE
best_alpha = None
best_mse = None

for alpha in alpha_values:
    reg = SGDRegressor(alpha=alpha, max_iter=1000, tol=1e-10)
    reg.fit(X, y)
    mse = mean_squared_error(y, reg.predict(X))

    # Печатаем текущее значение MSE
    print(f'Alpha: {alpha}, MSE: {mse}')

    # Считаем разницу между текущим MSE и целевым MSE
    mse_diff = abs(mse - target_mse)

    # Проверяем, насколько близко текущее значение MSE к целевому
    if mse_diff < best_mse_diff:
        best_mse_diff = mse_diff
        best_alpha = alpha
        best_mse = mse

print(f'Лучшее значение alpha: {best_alpha}, MSE: {best_mse}')
```

## Ваша многомерная линейная регрессия

***Задание 3 (5 баллов)***. Напишите собственную многомерную линейную регрессию, оптимизирующую MSE методом *градиентного спуска*. Для этого используйте шаблонный класс.

Критерий остановки: либо норма разности весов на текущей и предыдущей итерациях меньше определенного значения (первый и третий варианты), либо модуль разности функционалов качества (MSE) на текущей и предыдущей итерациях меньше определенного значения (второй и четвертый варианты). Также предлагается завершать обучение в любом случае, если было произведено слишком много итераций.

```python
class LinearRegressionGD:

    # Инициализация параметров модели
    def __init__(self, alpha=0.01, max_iter=1000, tol=1e-6):

        self.alpha = alpha                           # Скорость обучения (alpha)
        self.max_iter = max_iter                     # Максимальное количество итераций
        self.tol = tol                               # Порог для остановки (разность MSE)
        self.coefficients = None                     # Коэффициенты модели
        self.mse_history = []                        # История значений MSE для отладки



    # Обучение модели на данных X и y
    def fit(self, X, y):

        # param X: Матрица признаков (размер n_samples x n_features)
        # param y: Вектор целевой переменной (размер n_samples)

        n_samples, n_features = X.shape             # Получаем количество объектов и признаков

        # Инициализация коэффициентов случайными значениями
        self.coefficients = np.random.randn(n_features)

        # Основной цикл обучения
        for iteration in range(self.max_iter):
            # Предсказания модели (возвращаются предсказанные значения y)
            predictions = self.predict(X)

            # Вычисление ошибки (разность между предсказанными и истинными значениями)
            error = predictions - y

            # Вычисление MSE
            mse = np.mean(error ** 2)
            self.mse_history.append(mse)           # Сохраняем текущее значение MSE для дальнейшего анализа

            # Вывод информации о текущей итерации и значении MSE
            print(f'Итерация {iteration + 1}/{self.max_iter}, MSE: {mse:.6f}')

            # Обновление коэффициентов с помощью градиентного спуска
            gradient = (2 / n_samples) * X.T.dot(error)   # Вычисляем градиент MSE с помощью матричного умножения транспонированной матрицы признаков на вектор ошибок
            self.coefficients -= self.alpha * gradient    # Обновляем коэффициенты

            # Проверка условия остановки: изменение MSE меньше tol
            if iteration > 0 and abs(self.mse_history[-1] - self.mse_history[-2]) < self.tol:
                print(f'Остановка обучения на итерации {iteration + 1}: изменение MSE меньше {self.tol}')
                break                             # Прерываем обучение, если изменение MSE меньше заданного порога



    # Предсказание целевой переменной на основе признаков X
    def predict(self, X):

        # return: Вектор предсказанных значений (размер n_samples)
        return X.dot(self.coefficients)          # Возвращаем предсказанные значения



# Использование класса
# Генерация искусственных данных для обучения
X, y = make_regression(n_samples=1000, n_features=5, noise=0.1)

# Создание и обучение модели линейной регрессии с градиентным спуском
model = LinearRegressionGD(alpha=0.01, max_iter=1000, tol=1e-6)
model.fit(X, y)

# Вывод окончательных коэффициентов модели
print("Окончательные коэффициенты модели:", model.coefficients)
```

И тут я понял, что в предыдущем задании я, наверное, перемудрил...
Ладно, зато почти ничего менять не нужно. Только я проверку на условия остановки в цикле оставил внутри класса

```python
class LinearRegression(object):
    def __init__(self, alpha=0.0001, l_ratio=0.001, tol=0.001, max_iter=1000):
        '''
        Инициализация параметров
        alpha - скорость обучения
        l_ratio - параметр регуляризации (чем больше значение, тем сильнее будет штраф за большие коэффициенты)
        tol - значение для критерия останова
        max_iter - максимальное количество итераций обучения
        '''
        self.alpha = alpha                  # Скорость обучения
        self.l_ratio = l_ratio              # Параметр регуляризации
        self.tol = tol                      # Порог для остановки
        self.max_iter = max_iter            # Максимальное количество итераций
        self.coefficients = None            # Коэффициенты модели
        self.mse_history = []               # История значений MSE для отладки



    def fit(self, X, y):
        '''
        Метод для обучения линейной регрессии
        X - матрица признаков
        y - вектор правильных ответов
        '''

        n_samples, n_features = X.shape      # Получаем количество объектов и признаков

        # Инициализация коэффициентов случайными значениями
        self.coefficients = np.random.randn(n_features)

        # Основной цикл обучения
        for iteration in range(self.max_iter):
            # Предсказания модели
            predictions = self.predict(X)

            # Вычисление ошибки (разность между предсказанными и истинными значениями)
            error = predictions - y

            # Вычисление MSE
            mse = np.mean(error ** 2) + self.l_ratio * np.sum(self.coefficients ** 2)  # Добавляем L2 регуляризацию
            self.mse_history.append(mse)                # Сохраняем текущее значение MSE для дальнейшего анализа

            # Вывод информации о текущей итерации и значении MSE
            print(f'Итерация {iteration + 1}/{self.max_iter}, MSE: {mse:.6f}')

            # Обновление коэффициентов с помощью градиентного спуска с учетом L2 регуляризации
            gradient = (2 / n_samples) * X.T.dot(error) + 2 * self.l_ratio * self.coefficients  # Добавляем градиент L2 регуляризации
            self.coefficients -= self.alpha * gradient  # Обновляем коэффициенты

            # Проверка условия остановки: изменение MSE меньше tol = 1e-3
            if iteration > 0 and abs(self.mse_history[-1] - self.mse_history[-2]) < self.tol:
                print(f'Остановка обучения на итерации {iteration + 1}: изменение MSE меньше {self.tol}')
                break                                   # Прерываем обучение, если изменение MSE меньше заданного порога



    def predict(self, X):
        '''
        Метод для предсказаний линейной регрессии
        X - матрица признаков
        '''

        return X.dot(self.coefficients)                # Возвращаем предсказанные значения



# Использование класса
# Генерация искусственных данных для обучения (пример)
from sklearn.datasets import make_regression

X, y = make_regression(n_samples=1000, n_features=5, noise=0.1)

# Создание и обучение модели линейной регрессии с L2 регуляризацией
my_reg = LinearRegression(alpha=0.01, l_ratio=0.001)
my_reg.fit(X, y)

# Проверка качества модели с помощью MSE
mse_value = mean_squared_error(y, my_reg.predict(X))
print(f'Фактическое значение MSE: {mse_value}')       # Выводим значение MSE
```

***Задание 5 (1 балл)***. Обучите линейную регрессию из коробки

* с l1-регуляризацией (from sklearn.linear_model import Lasso, **первый и второй вариант**) или с l2-регуляризацией (from sklearn.linear_model import Ridge, **третий и четвертый вариант**)
* со значением параметра регуляризации **0.1 - для первого и третьего варианта, 0.01 - для второго и четвертого варианта**.

Обучите вашу линейную регрессию с тем же значением параметра регуляризации и сравните результаты. Сделайте выводы.

```python
from sklearn.linear_model import Ridge

# Ваша реализация линейной регрессии (с L2 регуляризацией)
class LinearRegression(object):
    def __init__(self, alpha=0.0001, l_ratio=0.01, tol=0.001, max_iter=1000):
        self.alpha = alpha                  # Скорость обучения
        self.l_ratio = l_ratio              # Параметр регуляризации
        self.tol = tol                      # Порог для остановки
        self.max_iter = max_iter            # Максимальное количество итераций
        self.coefficients = None            # Коэффициенты модели
        self.mse_history = []               # История значений MSE для отладки



    def fit(self, X, y):
        n_samples, n_features = X.shape     # Получаем количество объектов и признаков
        self.coefficients = np.random.randn(n_features)  # Инициализация коэффициентов

        for iteration in range(self.max_iter):
            predictions = self.predict(X)   # Вычисляем предсказания на основе текущих коэффициентов
            error = predictions - y         # Вычисляем ошибку предсказания

            # Вычисление MSE с учетом L2 регуляризации
            mse = np.mean(error ** 2) + self.l_ratio * np.sum(self.coefficients ** 2) # Вычисляем MSE с учетом L2 регуляризации
            self.mse_history.append(mse)    # Сохраняем текущее значение MSE

            # Обновление коэффициентов с учетом L2 регуляризации
            gradient = (2 / n_samples) * X.T.dot(error) + 2 * self.l_ratio * self.coefficients
            self.coefficients -= self.alpha * gradient

            # Проверка условия остановки: изменение MSE меньше tol
            if iteration > 0 and abs(self.mse_history[-1] - self.mse_history[-2]) < self.tol:
                break



    def predict(self, X):
        return X.dot(self.coefficients)    # Возвращаем предсказанные значения

# Генерация искусственных данных для обучения
X, y = make_regression(n_samples=1000, n_features=5, noise=0.1)

# Обучение линейной регрессии Ridge из коробки с L2-регуляризацией
ridge_model = Ridge(alpha=0.01)
ridge_model.fit(X, y)
ridge_predictions = ridge_model.predict(X)
ridge_mse = mean_squared_error(y, ridge_predictions)

# Обучение вашей реализации линейной регрессии с тем же значением параметра регуляризации
my_reg = LinearRegression(alpha=0.01, l_ratio=0.01)  # Установите l_ratio на 0.01
my_reg.fit(X, y)
my_reg_predictions = my_reg.predict(X)
my_reg_mse = mean_squared_error(y, my_reg_predictions)

# Сравнение результатов
print(f'MSE Ridge: {ridge_mse:.6f}')
print(f'MSE Custom Linear Regression: {my_reg_mse:.6f}')

# Выводы
if ridge_mse < my_reg_mse:
    print("Модель Ridge показала лучшее качество предсказания.")
elif ridge_mse > my_reg_mse:
    print("Ваша модель показала лучшее качество предсказания.")
else:
    print("Обе модели показали одинаковое качество предсказания.")
```

В отличие от обычной линейной регрессии, Ridge использует L2-регуляризацию, которая штрафует большие значения коэффициентов. Это помогает уменьшить их величину и тем самым улучшить обобщающую способность модели
