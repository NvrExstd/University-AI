1. Загрузка данных

```python
# 1.1(1.2) Импорт библиотек и датасета

import numpy as np

import pandas as pd

from matplotlib import pyplot as plt

all_data = pd.read_csv('groceries - groceries.csv')
all_data #Видно, что датафрейм содержит NaN значения
```

```python
# 1.3 Удаляем все значени NaN

np_data = all_data.to_numpy()
np_data = [[elem for elem in row[1:] if isinstance(elem,str)] for row in
np_data]
np_data
```

```python
# 1.4 Получаем список всех уникальных товаров

unique_items = set()
for row in np_data:
    for elem in row:
        unique_items.add(elem)

print('Количество уникальных позиций -',len(unique_items))

unique_items
```

```python
# 1.5 Выводим количество уникальных товаров
from collections import Counter

item_counts = Counter()

for row in np_data:
    item_counts.update(row)

print('Список товаров и их количество:')
for item, count in item_counts.items():
    print(f'{item}: {count}')

#print('Количество уникальных товаров -',len(unique_items))
```

2. FPGrowth и FPMax

```python
# 2.1 Преобразуем данные к виду, удобному для анализа

from mlxtend.preprocessing import TransactionEncoder

te = TransactionEncoder()
te_ary = te.fit(np_data).transform(np_data)
data = pd.DataFrame(te_ary, columns=te.columns_)
data

#Таблица представляет собой набор булевых значений напротив вида продукции по вертикали и напротив идентификатора покупателя по горизонтали.

# Ключевое отличие "data" от "np_data" заключается в том, что полученные данные представлены в виде таблицы, а исходные в виде списка транзакций, в котором каждая транзакция представлена списком товаров.

# Значение в таблице говорит о наличии конкретного вида продукции в корзине покупателя.
```

```python
# 2.2 Проведем ассоциативный анализ используя алгоритм FPGrowth при уровне поддержки 0.03

from mlxtend.frequent_patterns import fpgrowth
from IPython.core.display import display

results_gr = fpgrowth(data, min_support=0.03, use_colnames = True)
with pd.option_context('display.max_rows', None):
    display(results_gr)
results_gr.describe()

# В полученной таблице наблюдаем список товаров, которые встречаются в более чем 3% транзакций

# Каждая строка таблицы содержит информацию о наименования конкретной позиции и показатель "support", который указывает насколько часто встречается данная позиция. Умножив показатель "support" на 100, получим процентную величину
```

```python
# 2.3 Для анализа минимальных и максимальных значений для уровня поддержки наборов построим график, на котором обозначим точками эти значения

# Определяем диапазон уровней поддержки
min_support_range = np.arange(0.01, 0.4, 0.01)

# Создаем списки для хранения длин наборов и соответствующих поддержек
itemsets_lengths = []
threshold_supports = []
threshold_lengths = []

# Переменная для хранения последней длины наборов
last_itemset_len = len(data.columns)

# Проходим по каждому уровню поддержки в заданном диапазоне
for min_support in min_support_range:
    # Применяем алгоритм FPGrowth к данным с текущим уровнем поддержки
    results_gr = fpgrowth(data, min_support=min_support, use_colnames=True)

    # Сохраняем длину полученных наборов
    itemsets_lengths.append(len(results_gr))

    # Проверяем длину наборов и добавляем их в список
    if not results_gr.empty:
        results_gr['length'] = results_gr['itemsets'].apply(lambda x: len(x))
        current_itemset_max_len = results_gr['length'].max()

        # Проверяем на NaN и устанавливаем значение по умолчанию
        if np.isnan(current_itemset_max_len):
            current_itemset_max_len = 0

        # Если текущая максимальная длина меньше последней длины наборов,
        # обновляем значения порогов
        if current_itemset_max_len < last_itemset_len:
            last_itemset_len = current_itemset_max_len
            threshold_supports.append(min_support)
            threshold_lengths.append(len(results_gr))

# Строим график для визуализации результатов
plt.figure(figsize=(10, 6))
plt.plot(min_support_range.tolist(), itemsets_lengths, label='Количество наборов', marker='o')
plt.plot(threshold_supports, threshold_lengths, 'ro', label='Пороговые значения')

# Настраиваем график
plt.title('Количество наборов в зависимости от уровня поддержки (FPGrowth)')
plt.xlabel('Уровень поддержки')
plt.ylabel('Количество наборов')
plt.legend()
plt.grid()
plt.show()

# По полученному графику можно сделать следующий вывод: пороговые значения, отмеченные красными точками показывают минимальные и максимальные значения уровня поддержки для разных наборов товаров. 
```

```python
# 2.5 Сравним полученные результаты для FPGrowth и FPMax. 

results_gr = fpgrowth(data, min_support=0.03, use_colnames = True)
results_max = fpmax(data, min_support=0.03, use_colnames = True)
with pd.option_context('display.max_rows', None):
    display(results_gr.join(results_max, lsuffix='_gr', rsuffix='_max'))
```

```python
# Устанавливаем диапазон минимальной поддержки
min_support_range = np.arange(0.01, 0.4, 0.01)

# Создаем списки для хранения длин наборов и соответствующих поддержек для FPGrowth
itemsets_lengths_fpgrowth = []
threshold_supports_fpgrowth = []
threshold_lengths_fpgrowth = []

# Переменная для хранения последней длины наборов
last_itemset_len = len(data.columns)

# Проходим по каждому уровню поддержки в заданном диапазоне для FPGrowth
for min_support in min_support_range:
    results_gr = fpgrowth(data, min_support=min_support, use_colnames=True)
    itemsets_lengths_fpgrowth.append(len(results_gr))

    if not results_gr.empty:
        results_gr['length'] = results_gr['itemsets'].apply(lambda x: len(x))
        current_itemset_max_len = results_gr['length'].max()

        if np.isnan(current_itemset_max_len):
            current_itemset_max_len = 0

        if current_itemset_max_len < last_itemset_len:
            last_itemset_len = current_itemset_max_len
            threshold_supports_fpgrowth.append(min_support)
            threshold_lengths_fpgrowth.append(len(results_gr))

# Создаем списки для хранения длин наборов и соответствующих поддержек для FPMax
itemsets_lengths_fpmax = []
threshold_supports_fpmax = []
threshold_lengths_fpmax = []

# Сброс переменной для хранения последней длины наборов
last_itemset_len = len(data.columns)

# Проходим по каждому уровню поддержки в заданном диапазоне для FPMax
for min_support in min_support_range:
    results_max = fpmax(data, min_support=min_support, use_colnames=True)
    itemsets_lengths_fpmax.append(len(results_max))

    if not results_max.empty:
        results_max['length'] = results_max['itemsets'].apply(lambda x: len(x))
        current_itemset_max_len = results_max['length'].max()

        if np.isnan(current_itemset_max_len):
            current_itemset_max_len = 0

        if current_itemset_max_len < last_itemset_len:
            last_itemset_len = current_itemset_max_len
            threshold_supports_fpmax.append(min_support)
            threshold_lengths_fpmax.append(len(results_max))

# Строим график для визуализации результатов
plt.figure(figsize=(10, 6))

# График для FPGrowth
plt.plot(min_support_range.tolist(), itemsets_lengths_fpgrowth, label='FPGrowth: Количество наборов', marker='o')
plt.plot(threshold_supports_fpgrowth, threshold_lengths_fpgrowth, 'ro', label='FPGrowth: Пороговые значения')

# График для FPMax
plt.plot(min_support_range.tolist(), itemsets_lengths_fpmax, label='FPMax: Количество наборов', marker='x')
plt.plot(threshold_supports_fpmax, threshold_lengths_fpmax, 'go', label='FPMax: Пороговые значения')

# Настраиваем график
plt.title('Количество наборов в зависимости от уровня поддержки')
plt.xlabel('Уровень поддержки')
plt.ylabel('Количество наборов')
plt.legend()
plt.grid()
plt.show()
```

```python
# Разница между FPMax и FPGrowth заключается в том, что алгоритм FPMax выводит только "максимальные" наборы. 
# Максимальный набор - это набор, который является частым, то есть имеет уровень поддержки больше, чем минимальная граница, и при этом не является поднабором для другого частого набора.

# Из полученных таблицы и графика можно пронублюдать, что FPMax действительно делает выборку меньшего количества наборов.
```

```python
# 2.9 Построим график изменения количества получаемых правил от уровня поддержки. 
# Из графика видно, что полученные кривые кореллируют с кривыми, полученными ранее на других графиках. 
# Точки пересечения на графиках означают места минимальные и максимальные уровни поддержки для разных наборов товаров
# Похожие участки кривых между точками пересечений наблюдаются на графиках, полученных ранее, которые отражали общую картину

# Определяем диапазон уровней поддержки
min_support_range = np.arange(0.01, 0.4, 0.01)

# Создаем словарь для хранения количества правил для каждого размера наборов
rules_count = {i: [] for i in range(1, 5)}             # Предполагаем, что нас интересуют наборы до размера 4

# Проходим по каждому уровню поддержки в заданном диапазоне
for min_support in min_support_range:
    # Применяем алгоритм FPGrowth к данным с текущим уровнем поддержки
    results_gr = fpgrowth(data, min_support=min_support, use_colnames=True)

    # Сохраняем количество правил для каждого размера наборов
    for size in range(1, 5):  # От 1 до 4
        # Фильтруем результаты по размеру наборов
        count = results_gr['itemsets'].apply(lambda x: len(x) == size).sum()
        rules_count[size].append(count)

# Построение графика
plt.figure(figsize=(20, 10))

# Отображаем кривые для каждого размера наборов товаров
for size in range(1, 5):
    plt.plot(min_support_range.tolist(), rules_count[size], marker='o', label=f'Наборы из {size} товаров')

# Настраиваем график
plt.title('Изменение количества правил от уровня поддержки')
plt.xlabel('Уровень поддержки')
plt.ylabel('Количество правил')
plt.xticks(min_support_range.tolist())                 # Устанавливаем метки по оси X
plt.legend()
plt.grid()
plt.show()
```

```python
# 3.3  Проведем ассоциативный анализ

# antecedents           - Это набор товаров, который является предшествующим элементом в правиле.
#
# consequents           - Это набор товаров, который является последующим элементом в правиле.
#
# antecedent support    - Это мера того, насколько часто антецедент встречается в данных.
#
# consequent support    - Это мера того, насколько часто консеквент встречается в данных.
#
# support               - Это мера того, насколько часто антецедент и консеквент встречаются вместе в данных. Поддержка рассчитывается как доля транзакций, содержащих как антецедент, так и консеквент.
#
# confidence            - Это мера того, насколько вероятно, что консеквент будет куплен, если антецедент уже куплен. 
#                         Она рассчитывается как отношение числа транзакций, содержащих как антецедент, так и консеквент, к числу транзакций, содержащих только антецедент.
#
# lift                  - Это мера того, насколько антецедент и консеквент зависят друг от друга. 
#                         Lift рассчитывается как отношение фактической вероятности совместного появления антецедента и консеквента к вероятности их независимого появления. 
#                         Lift больше 1 указывает на положительную зависимость, меньше 1 — на отрицательную.
#
# leverage              - Это мера того, насколько вероятность совместного появления антецедента и консеквента превышает ожидаемую вероятность, если бы они были независимыми. 
#                         Рассчитывается как разница между фактической поддержкой и ожидаемой поддержкой, если бы антецедент и консеквент были независимыми. Положительное значение указывает на положительную зависимость.
#
# conviction            - Это мера того, насколько антецедент предсказывает консеквент. Рассчитывается как отношение вероятности того, что антецедент произойдет, к вероятности того, что консеквент не произойдет, при условии, что антецедент произошел. 
#                         Чем выше значение, тем сильнее предсказательная сила антецедента.
#
# zhangs_metric         - Эта метрика измеряет степень зависимости между антецедентом и консеквентом. Она учитывает как поддержку, так и доверие, и может быть полезной для выявления сильных ассоциаций между товарами.
#                         Значение больше 1: Указывает на положительную зависимость между антецедентом и консеквентом. Это означает, что наличие антецедента увеличивает вероятность покупки консеквента.
#                         Значение равно 1: Указывает на независимость между антецедентом и консеквентом. Это означает, что антецедент и консеквент не влияют друг на друга.
#                         Значение меньше 1: Указывает на отрицательную зависимость. Это означает, что наличие антецедента снижает вероятность покупки консеквента.

from mlxtend.frequent_patterns import association_rules

rules = association_rules(result, min_threshold = 0.3)
rules

# 3.4 Согласно документации, по умолчанию используется расчет на основе метрики 'confidence'
```

```python
# 3.7 Построим граф для следующего анализа

rules = association_rules(result, metric='confidence', min_threshold = 0.4)
rules

#antecedents - левая часть правила (предшествующие элементы). Это элементы, которые должны присутствовать в транзакции, чтобы выполнялось правило.

#consequents - правая часть правила (следствующие элементы). Это элементы, которые должны присутствовать в транзакции, если выполняется левая часть правила.
```

```python
import networkx as nx
import matplotlib.pyplot as plt

plt.figure(figsize=(16, 9))

G = nx.DiGraph()

for i in range (rules.shape[0]):
    antecedent = ', '.join(rules.iloc[i].antecedents)
    consequent = ', '.join(rules.iloc[i].consequents)
    support = rules.iloc[i].support
    confidence = rules.iloc[i].confidence

    G.add_edge(antecedent, consequent, support=support, confidence=confidence)

edge_widths = [G[u][v]['support'] * 100 for u, v in G.edges()]  

edge_colors = [G[u][v]['confidence'] for u, v in G.edges()]
norm = plt.Normalize(min(edge_colors), max(edge_colors))
cmap = plt.cm.coolwarm
edge_colors = [cmap(norm(color)) for color in edge_colors]

edge_labels = {(u, v): f'conf: {G[u][v]["confidence"]:.2f}' for u, v in G.edges()}

pos = nx.spring_layout(G) 
nx.draw(G, pos, with_labels=True, node_size=2000, node_color='lightblue', font_size=10, font_weight='bold', width=edge_widths, edge_color=edge_colors)
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels)

plt.title('Граф ассоциативных правил')

plt.show()
```

```python
# 3.9 Как дополнительный способ визуализации можно было бы использовать тепловую карту.
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from mlxtend.frequent_patterns import association_rules

# Предполагаем, что 'result' - это DataFrame с частыми наборами
rules = association_rules(result, metric='confidence', min_threshold=0.4)

# Создаем DataFrame для тепловой карты
# Для этого нам нужно создать матрицу, где строки - это антецеденты, а столбцы - это последующие
# Мы будем использовать confidence в качестве значений

# Создаем пустую матрицу
matrix = pd.DataFrame(0, index=rules['antecedents'].apply(lambda x: ', '.join(list(x))), 
                      columns=rules['consequents'].apply(lambda x: ', '.join(list(x))))

# Заполняем матрицу значениями confidence
for index, row in rules.iterrows():
    antecedent = ', '.join(list(row['antecedents']))
    consequent = ', '.join(list(row['consequents']))
    matrix.loc[antecedent, consequent] = row['confidence']

# Настройка графика
plt.figure(figsize=(12, 8))
sns.heatmap(matrix, annot=True, cmap='YlGnBu', cbar=True, fmt=".2f", linewidths=.5)

# Настраиваем график
plt.title('Тепловая карта правил ассоциации (Confidence)')
plt.xlabel('Последующие товары')
plt.ylabel('Антецеденты')
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.tight_layout()

# Отображаем график
plt.show()
```
