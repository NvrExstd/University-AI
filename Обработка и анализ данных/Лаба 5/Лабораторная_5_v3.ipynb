**1. Загрузка данных**

1.1 Загрузить датасет по ссылке: https://archive.ics.uci.edu/dataset/109/wine.

1.2 Создать Python скрипт. Загрузить данные в датафрейм

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler

column_names = [
    'Class', 'Alcohol', 'Malic_Acid', 'Ash', 'Alcalinity_of_Ash',
    'Magnesium', 'Total_Phenols', 'Flavanoids', 'Nonflavanoid_Phenols',
    'Proanthocyanins', 'Color_Intensity', 'Hue',
    'OD280_OD315', 'Proline'
]

df = pd.read_csv('wine.data', header=None, names=column_names)

print("Данные о винах:")
df
```

1.3. При необходимости, произвести стандартизацию данных с
использованием preprocessing.StandardScaler() из sklearn

```python
scaler = StandardScaler()

scaled_features = scaler.fit_transform(df)

scaled_df = pd.DataFrame(scaled_features, columns=df.columns)

scaled_df['Class'] = df['Class']

print("\nСтандартизированные данные о винах:")
scaled_df
```

1.4. Подготовить данные для дальнейшего анализа:

*   Понизить размерность пространства данных до размерности n , при
которой компоненты объясняют не менее 85% дисперсии данных.
Понижение размерности пространства осуществляется с помощью
метода главных компонент (PCA.fit_transform)
*   Восстановить данные для полученного количества компонент
(PCA.inverse_transform).



```python
from sklearn.decomposition import PCA

pca = PCA(n_components=0.85, svd_solver='full')

reduced_data = pca.fit_transform(scaled_df.iloc[:, 1:])

reconstructed_data = pca.inverse_transform(reduced_data)

reconstructed_df = pd.DataFrame(reconstructed_data, columns=scaled_df.columns[1:])

reconstructed_df.insert(0, 'Class',scaled_df['Class'].values)

print("\nВосстановленные данные после понижения размерности:")
reconstructed_df
```

**2. K-Mean**

2.1. Провести кластеризацию методом k-средних. Вывести на экран
значения центров кластеров и количество наблюдений, которое попало в
каждый кластер (Использовать: KMeans из sklearn.cluster)

```python
from sklearn.cluster import KMeans
import numpy as np

n_clusters = 3

kmeans = KMeans(n_clusters=n_clusters, random_state=42)

kmeans.fit(reduced_data)

cluster_centers = kmeans.cluster_centers_

labels = kmeans.labels_

unique, counts = np.unique(labels, return_counts=True)
cluster_counts = dict(zip(unique, counts))

print("Центры кластеров:")
print(cluster_centers)

print("\nКоличество наблюдений в каждом кластере:")
for cluster, count in cluster_counts.items():
    print(f"Кластер {cluster}: {count} наблюдений")
```

2.2. Построить (графически) результаты классификации для признаков
попарно (1 и 2, 2 и 3, ..., n -1 и n ), отобразить центры кластеров

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="darkgrid")

n_features = reduced_data.shape[1]

n_rows = 3
n_cols = (n_features - 1) // n_rows + 1 

fig, axes = plt.subplots(n_rows, n_cols, figsize=(20, 15))
axes = axes.flatten()

def plot_scatter(ax, x, y, cluster_centers, i, j):
    sns.scatterplot(x=x, y=y, hue=labels, palette='Accent', ax=ax, s=50)
    ax.scatter(cluster_centers[:, i], cluster_centers[:, j], color='Red', marker='X', s=200, label='Центры кластеров')
    ax.set_title(f'Компоненты: {i + 1} и {j + 1}')
    ax.set_xlabel(f'Компонента {i + 1}')
    ax.set_ylabel(f'Компонента {j + 1}')
    ax.legend()

index = 0
for i in range(n_features - 1):
    j = i + 1
    plot_scatter(axes[index], reduced_data[:, i], reduced_data[:, j], cluster_centers, i, j)
    index += 1

for i in range(index, len(axes)):
    fig.delaxes(axes[i])

plt.tight_layout()
plt.show()
```

Дать пояснения:
1. На что влияет параметр n_init ?
2. Чему равно оптимальное значение для n_init ? Почему это
оптимальное значение?
3. Что происходит при увеличении и уменьшении параметра n_init от
оптимального?
4. По каким признакам произошло наилучшее разделение?
5. Как изменятся результаты, если в качестве метода инициализации
выбрать random ?

```python
# 1. На что влияет параметр n_init ?

n_init_values = range(1, 21)
inertia_values = []

for n_init in n_init_values:
    kmeans = KMeans(n_clusters=3, n_init=n_init, random_state=42)
    kmeans.fit(reduced_data)
    inertia_values.append(kmeans.inertia_)
    print(f"n_init={n_init}, Inertia={kmeans.inertia_:.2f}")

plt.figure(figsize=(10, 6))
plt.plot(n_init_values, inertia_values, marker='o', markersize=10, linewidth=3, color='red')
plt.title('Влияние параметра n_init на инерцию')
plt.xlabel('n_init')
plt.ylabel('Инерция')
plt.xticks(n_init_values)
plt.grid(True)
plt.show()

# Параметр n_init в методе k-средних влияет на стабильность и качество конечного результата кластеризации. 
# Данный параметр задает количество повторений алгоритма k-средних, в котором он будет инициализироваться с различными начальными точками перед выбором наилучшего результата, а именно с минимальным значением инерции. 
# Инерция (inertia) отражает суммарное квадратичное расстояние точек от центров их кластеров. Кучность кластеров окажется сильнее при меньших значениях инерции.
 
# Из графика видно, что при n_init=1 инерция равна 934.68 и не изменяется при следующих итерациях. 
# Это означает, что начиная с n_init=1, алгоритм смог найти оптимальное разбиение с полученным значением инерциеи. 
# Можно сделать вывод, что увеличение n_init не влияет на результаты кластеризации.
```

```python
# 2. Чему равно оптимальное значение для n_init ? Почему это оптимальное значение?

min_inertia = min(inertia_values)
optimal_n_init = n_init_values[inertia_values.index(min_inertia)]

print(f"\nОптимальное значение n_init: {optimal_n_init}")
print(f"Минимальная инерция: {min_inertia:.2f}")

# Ответ на данный вопрос исходит из предыдущего пункта, оптимальное значение n_init = 1. Нет смысла повторно запускать алгоритм, эффективность не изменяется.
```

```python
# 3. Что происходит при увеличении и уменьшении параметра n_init от оптимального?

# Теоретически, при меньшем n_init должно быть менее оптимальное разбиение. 
# При меньшем количестве инициализаций вероятность получить менее удачное начальное положение центроидов возрастает, что приводит к менее компактным кластерам.
 
# При бОльшем n_init инерция остается на том же уровне, что совпадает с минимальным значением. 
# Дальнейшее увеличение n_init не улучшает результат, так как алгоритм уже нашел оптимальное разбиение.
 
# На наших данных от увеличения или уменьшения n_init оптимальность разбиения не изменяется, что указывает на удачную кластеризацию с 1 повторения алгоритма.
```

```python
# 4. По каким признакам произошло наилучшее разделение?

# По графикам из пункта 2.2 видно, что наилучшее разделение произошло между 1 и 2 компонентами. Центроиды достаточно далеко друг от друга, точки кластера чётко разделены между собой.
```

```python
new_data = PCA(n_components=2).fit_transform(reduced_data) 

n_clusters = 3

kmeans = KMeans(init='k-means++', n_clusters=n_clusters, n_init=4)

kmeans.fit(new_data)

h = 0.02

x_min, x_max = new_data[:, 0].min() - 1, new_data[:, 0].max() + 1
y_min, y_max = new_data[:, 1].min() - 1, new_data[:, 1].max() + 1

xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))

sk_kmeans_pred_res = kmeans.predict(new_data)

for i in range(n_clusters):
    count_of_clusters = np.sum(sk_kmeans_pred_res == i)
    print('В кластере №', i, count_of_clusters, 'элементов')

Z = kmeans.predict(np.c_[xx.ravel(), yy.ravel()])

Z = Z.reshape(xx.shape)
plt.figure(1, figsize=(15, 10))
plt.clf()
plt.imshow(
    Z,
    interpolation="nearest",
    extent=(xx.min(), xx.max(), yy.min(), yy.max()),
    cmap=plt.cm.Paired,
    aspect="auto",
    origin="lower",
)

plt.scatter(new_data[:, 0], new_data[:, 1], c=kmeans.labels_, cmap='autumn', s=60, edgecolor='k')

plt.plot(new_data[:, 0], new_data[:, 1], "k.", markersize=2)

centroids = kmeans.cluster_centers_
print("Центры", centroids, '', sep='\n')
plt.scatter(
    centroids[:, 0],
    centroids[:, 1],
    marker="x",
    s=169,
    linewidths=3,
    color="w",
    zorder=10,
)

plt.title(
    "Кластеризация k-средних на уменьшенных данных\n(данные, уменьшенные PCA)\n"
    "Центроиды отмечены белым крестом"
)
plt.xlim(x_min, x_max)
plt.ylim(y_min, y_max)
plt.xticks(())
plt.yticks(())

plt.show()
```

2.4 Исследуйте работу алгоритма k-средних при различных параметрах init.

```python
init_options = [
    ('random', 1, 3),  # Случайная инициализация, 1 запуск, максимум 10 итераций
    ('random', 5, 10),  # Случайная инициализация, 5 запусков, максимум 50 итераций
    ('random', 10, 200),   # Случайная инициализация, 10 запусков, максимум 200 итераций
    ('k-means++', 1, 3),  # Инициализация k-means++, 1 запуск, максимум 10 итераций
    ('k-means++', 5, 10),  # Инициализация k-means++, 5 запусков, максимум 50 итераций
    ('k-means++', 10, 200)  # Инициализация k-means++, 10 запусков, максимум 200 итераций
]

fig, axes = plt.subplots(2, 3, figsize=(15, 10))
axes = axes.flatten() 

for i, (init, n_init, max_iter) in enumerate(init_options):

    kmeans = KMeans(n_clusters=3, init=init, n_init=n_init, max_iter=max_iter, random_state=42)

    kmeans.fit(new_data)

    axes[i].scatter(new_data[:, 0], new_data[:, 1], c=kmeans.labels_, cmap='autumn', s=30)

    axes[i].scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=100, c='Blue', marker='X', label='Центры кластеров')

    title = f"init: {init}, n_init: {n_init}, max_iter: {max_iter}"
    axes[i].set_title(title)

plt.tight_layout()

plt.show()
```

2.6 Проведите кластеризацию используя пакетную кластеризацию k-
средних. Постройте диаграмму рассеяния, на которой будут выделено точки,
которые для разных методов попали в разные кластеры

```python
from sklearn.cluster import MiniBatchKMeans

n_clusters = 3 

kmeans = KMeans(n_clusters=n_clusters, init='k-means++', n_init=10, max_iter=300, random_state=42)
kmeans_labels = kmeans.fit_predict(new_data) 

minibatch_kmeans = MiniBatchKMeans(n_clusters=n_clusters, batch_size=100, random_state=42)
minibatch_labels = minibatch_kmeans.fit_predict(new_data)

plt.figure(figsize=(20, 7))

plt.subplot(1, 3, 1) 
plt.scatter(new_data[:, 0], new_data[:, 1], c=kmeans_labels, cmap='autumn', s=30, edgecolor='k')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], marker='X', s=200, c='red', label='Центры кластеров')
plt.title('Кластеризация KMeans')
plt.xlabel('Компонента 1')
plt.ylabel('Компонента 2')
plt.legend()

plt.subplot(1, 3, 2)
plt.scatter(new_data[:, 0], new_data[:, 1], c=minibatch_labels, cmap='autumn', s=30, edgecolor='k')
plt.scatter(minibatch_kmeans.cluster_centers_[:, 0], minibatch_kmeans.cluster_centers_[:, 1], marker='X', s=200, c='blue', label='Центры кластеров')
plt.title('Кластеризация MiniBatchKMeans')
plt.xlabel('Компонента 1')
plt.ylabel('Компонента 2')
plt.legend()

plt.subplot(1, 3, 3)
for i in range(n_clusters):

    kmeans_points = new_data[kmeans_labels == i]
    minibatch_points = new_data[minibatch_labels == i]

    plt.scatter(kmeans_points[:, 0], kmeans_points[:, 1], color='lightgray', alpha=0.5)

    plt.scatter(minibatch_points[:, 0], minibatch_points[:, 1], color='lightgray', alpha=0.5)

for idx in range(len(new_data)):
    if kmeans_labels[idx] == minibatch_labels[idx]:
        plt.scatter(new_data[idx][0], new_data[idx][1], color='lightgray', alpha=0.5)
    else:
        plt.scatter(new_data[idx][0], new_data[idx][1], color='red', alpha=0.5)

centroids_kmeans = kmeans.cluster_centers_
centroids_minibatch = minibatch_kmeans.cluster_centers_
plt.scatter(centroids_kmeans[:, 0], centroids_kmeans[:, 1], marker='X', s=200, c='red', label='Центры KMeans')
plt.scatter(centroids_minibatch[:, 0], centroids_minibatch[:, 1], marker='X', s=200, c='blue', label='Центры MiniBatch')

plt.title('Сравнение меток кластеров между KMeans и MiniBatchKMeans')
plt.xlabel('Компонента 1')
plt.ylabel('Компонента 2')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
# 1. Пакетная кластеризация k-средних отличается от обычного метода k-средних тем, что она обрабатывает данные не целиком, а использует небольшие случайные подвыборки (пакеты) данных на каждой итерации. 
# Это значительно сокращает время вычислений и снижает потребление памяти. Это наиболее ощутимо при работе с большими наборами данных. 
# В результате пакетная кластеризация может быть более быстрой, но иногда менее точной, чем классический метод k-средних, поскольку обновление центров кластеров происходит на основе меньшего объема данных.
# 
# 2. По графикам можно сделать вывод, что у метода Kmeans нет четкой границы у желтого и фиолетового кластера, в то время как у метода MiniBatchKMeans у данных кластеров граница куда более четкая. 
# Но у зеленого и фиолетового кластеров разбиение оказалось хуже чем у KMeans. На третьем графике представлено сравнение меток кластеров между Kmeans и MiniBatchKmeans. 
# Бледно-серым цветом отмечены совпадающие метки, красным - несовпадающие. На графике хорошо заметны несовпадающие метки у желтого и фиолетового кластеров. Более того, они поменялись местами, но это связано с различием работы методов. Значительно смещены центроиды кластеров.
```

**3. Иерархическая кластеризация**

3.1 Провести и отобразить иерархическую кластеризацию на тех же данных (см. п.1.4. и п.2.2) с параметром average. Использовать: AgglomerativeClustering из sklearn.cluster

```python
from sklearn.cluster import AgglomerativeClustering

n_clusters = 3 

kmeans = KMeans(n_clusters=n_clusters, init='k-means++', n_init=10, max_iter=300, random_state=42)
kmeans_labels = kmeans.fit_predict(new_data)

agglo = AgglomerativeClustering(n_clusters=n_clusters, linkage='average')
agglo_labels = agglo.fit_predict(new_data)

plt.figure(figsize=(20, 7))

plt.subplot(1, 3, 1)
plt.scatter(new_data[:, 0], new_data[:, 1], c=kmeans_labels, cmap='autumn', s=30, edgecolor='k')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], marker='X', s=200, c='blue', label='Центры кластеров')
plt.title('Кластеризация KMeans')
plt.xlabel('Компонента 1')
plt.ylabel('Компонента 2')
plt.legend()

plt.subplot(1, 3, 2)
plt.scatter(new_data[:, 0], new_data[:, 1], c=agglo_labels, cmap='autumn', s=30, edgecolor='k')
plt.title('AgglomerativeClustering')
plt.xlabel('Компонента 1')
plt.ylabel('Компонента 2')

plt.subplot(1, 3, 3)
for i in range(n_clusters):

    kmeans_points = new_data[kmeans_labels == i]
    agglo_points = new_data[agglo_labels == i]

    plt.scatter(kmeans_points[:, 0], kmeans_points[:, 1], color='lightgray', alpha=0.5)

    plt.scatter(agglo_points[:, 0], agglo_points[:, 1], color='lightgray', alpha=0.5)

for idx in range(len(new_data)):
    if kmeans_labels[idx] == agglo_labels[idx]:
        plt.scatter(new_data[idx][0], new_data[idx][1], color='lightgray', alpha=0.5) 
    else:
        plt.scatter(new_data[idx][0], new_data[idx][1], color='red', alpha=0.5)

centroids_kmeans = kmeans.cluster_centers_
plt.scatter(centroids_kmeans[:, 0], centroids_kmeans[:, 1], marker='X', s=200, c='blue', label='Центры KMeans')

plt.title('Сравнение меток кластеров между KMeans и AgglomerativeClustering')
plt.xlabel('Компонента 1')
plt.ylabel('Компонента 2')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
# 1. Первый график показывает результаты кластеризации методом KMeans. Он разбивает данные на три кластера, определяя центры кластеров (синие кресты). 
# Второй график показывает результаты кластеризации методом AgglomerativeClustering с параметром average. На третьем графике видно, что некоторые точки, относящиеся к одному кластеру в KMeans, отнесены к другому кластеру в AgglomerativeClustering (отмеченные красным цветом). 
# По первому графику можно заметить, что есть нечеткое разделение желтого и красного кластера, однако со зрительной точки зрения разбиение кажется более логичным, потому как точки KMeans определены к кластерам ближе к центроидам. 
# На графике AgglomerativeClustering границы более ровные, но распределение выглядит менее логичным. Это связано с разницей в алгоритмах работы методов. 
# Из графика видно, что красный и оранжевый кластеры поменялись местами, из-за чего на третьем графике метки двух кластеров полностью красные. 
# Смена меток кластеров в AgglomerativeClustering по сравнению с KMeans объясняется различной логикой формирования кластеров:
#                            KMeans начинает с инициализации центров кластеров, далее точки распределяются вокруг этих центров
#                            AgglomerativeClustering работает иначе: он последовательно объединяет ближайшие точки или группы точек. 
# Порядок, в котором точки объединяются в кластеры, не зависит от начального положения центров, это может привести к иному распределению по кластерам.

# 2. KMeans показал себя более эффективным в данном случае, группы выделены более чётко. Кластеры компактнее, легко идентифицируются.
```

3.4 Сгенерируйте случайные данные (x,y) в виде двух квадратных контуров

Замечание:
* Общее количество точек 750: внешний квадратный контур - 500; внутренний квадратный контур - 250.
* Проведите их иерархическую кластеризацию со всеми возможными параметрами linkage.
* Отобразите полученные результаты

```python
np.random.seed(42)

# Внешний квадрат (500 точек)
outer_square = np.vstack([
    np.random.uniform(-10, 10, size=(125, 2)),  # Левая сторона
    np.random.uniform(-10, 10, size=(125, 2)),  # Правая сторона
    np.random.uniform(-10, 10, size=(125, 2)),  # Верхняя сторона
    np.random.uniform(-10, 10, size=(125, 2))   # Нижняя сторона
])

# Внешний квадрат (250 точек)
inner_square = np.vstack([
    np.random.uniform(-5, 5, size=(63, 2)),   # Левая сторона
    np.random.uniform(-5, 5, size=(62, 2)),   # Правая сторона
    np.random.uniform(-5, 5, size=(62, 2)),   # Верхняя сторона
    np.random.uniform(-5, 5, size=(63, 2))    # Нижняя сторона
])

X = np.vstack([outer_square, inner_square])

linkage_methods = ['ward', 'complete', 'average', 'single']

fig, axes = plt.subplots(1, 4, figsize=(28, 6))
fig.suptitle("Иерархическая кластеризация с различными методами linkage")

for i, method in enumerate(linkage_methods):
    clustering = AgglomerativeClustering(n_clusters=2, linkage=method) 
    labels = clustering.fit_predict(X) 

    axes[i].scatter(X[:, 0], X[:, 1], c=labels, cmap='autumn', s=10)
    axes[i].set_title(f"{method}")
    axes[i].set_xticks([])  
    axes[i].set_yticks([]) 

plt.show()
```

Дать пояснения:
1. Какой тип связи работает лучше всего и почему?

```python
# 1. Linkage: ward
# Метод ward минимизирует сумму квадратов отклонений внутри кластеров, что делает его пригодным для создания плотных и компактных кластеров. 
# По графику можно сделать вывод, что данный метод показал четкое разделение на два кластера. Хорошо просматривается граница разделения данных по центру.
# 
# 2. Linkage: complete
# Данный метод использует максимальное расстояние между точками кластеров, что может способствовать образованию более четких границ между кластерами. 
# Большинство точек распределено равномерно по кластерам, но граница стала менее чёткой.
# 
# 3. Linkage: average
# Метод average показал очень схожий результат с ward и complete. Кластеры здесь также не выделяют четко внешний и внутренний квадраты. 
# Поскольку average linkage использует среднее расстояние между точками кластеров, полученный результат выглядит достаточно хорошим, граница наиболее широкая по сравнению с остальными методами.
# 
# 4. Linkage: single
# Метод single дал самый слабый результат: практически все точки объединены в один кластер, за исключением нескольких точек, которые были выделены отдельно.
# Single linkage основан на минимальном расстоянии между точками кластеров, и это делает его склонным к образованию вытянутых, цепочечных кластеров. В данном случае он не сумел правильно выделить кластеры для обоих контуров, так как данные распределились в одном основном кластере.
# 
# В данной ситуации наиболее предпочтительным является метод Ward, так как разделение на кластеры оказалось наиболее равномерным, граница относительно ровная, размеры кластеров приблизительно идентичны друг другу. 
# Метод Average в этом случае уступает, хоть и вид границы более предпочтителен, но из графика видно, что фиолетовый кластер начинает преобладать.
```
